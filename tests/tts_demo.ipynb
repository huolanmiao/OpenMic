```python
text = [t+'[lbreak]' for t in text]

merge_sentences = False

if merge_sentences:
    text = ''.join(text)
```

```python

params_infer_code = ChatTTS.Chat.InferCodeParams(
    prompt="[speed_2]",
    top_K=15,
    top_P=0.7,
    temperature=0.4,
    spk_emb=rand_spk,
    max_new_token=4096
)
params_refine_text = ChatTTS.Chat.RefineTextParams(
    prompt="[oral_0][laugh_1][break_4]",
    max_new_token=2048
)

wav = chat.infer(
    text,
    params_refine_text=params_refine_text,
    params_infer_code=params_infer_code,
)
```
